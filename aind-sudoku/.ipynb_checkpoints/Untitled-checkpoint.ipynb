```python
def cross(A, B):
    "Cross product of elements in A and elements in B."
    C = [x + y for x in A for y in B]
    return C
rows = 'ABCDEFGHI'
cols = '123456789'
D = cross(rows, cols)
print(D)
for z in cols:
    print(z)
    
boxes = cross(rows, cols)

row_units = [cross(r, cols) for r in rows]
column_units = [cross(rows, c) for c in cols]
square_units = [cross(rs, cs) for rs in ('ABC','DEF','GHI') for cs in ('123','456','789')]
diagonalunits = [[s+t for s,t in zip(rows,cols)]]
diagonalunitsreversed = [[s+t for s,t in zip(rows,reversed(cols))]]
unitlist = row_units + column_units + square_units + diagonalunits + diagonalunitsreversed

units = dict((s, [u for u in unitlist if s in u]) for s in boxes)
peers = dict((s, set(sum(units[s],[]))-set([s])) for s in boxes)
print(units)
```

```python
from collections import OrderedDict
def grid_values(grid):
    """
    Convert grid into a dict of {square: char} with '123456789' for empties.
    Args:
        grid(string) - A grid in string form.
    Returns:
        A grid in dictionary form
            Keys: The boxes, e.g., 'A1'
            Values: The value in each box, e.g., '8'. If the box has no value, then the value will be '123456789'.
    """
    rows = 'ABCDEFGHI'
    cols = '123456789'
    boxes = cross(rows,cols)
    board = {}
    for x,box in zip(grid,boxes):
        if x == '.':
            board[box] = '123456789'
        else:
            board[box] = x

    
    return board

def display(values):
    """
    Display the values as a 2-D grid.
    Args:
        values(dict): The sudoku in dictionary form
    """
    
    D = OrderedDict(sorted(values.items()))
   
    

    nextkey,_ = next(iter(D.items()))
    next_key = nextkey[:1] 
    #print(next_key)
    lines = []
    line = []
    len_dict = len(D)
    no_key = 0
    for (key) in D:
        #print(key)
        no_key +=1
        str1 = D[key].rjust(9)
   
        if (key[:1] == next_key):        
            if(no_key%3 == 0):
                line.append(str1.join(' |'))
            else:
                line.append(str1.join('  '))
            if (no_key == len_dict):
                lines.append(line.copy())
        else:        
            next_key = key[:1]
            lines.append(line.copy())
        
            if((no_key-1) % 27 == 0):
                lines.append('+'.join(['-'*(11*3 - 1)]*3)) 
            line[:]=[]
            if(no_key%3 == 0):
                line.append(str1.join(' |'))
            else:
                line.append(str1.join('  '))
    

    for lin in lines:
        for l in lin:
            print(l,end="")
        print()
grid = '2.............62....1....7...6..8...3...9...7...6..4...4....8....52.............3'        
board = grid_values(grid)
display(board)
```

```python

def eliminateNakedTwin(values,twins):
        print("+++++++")
        print(values)
        print(twins)
        print("----------")
        peers_box = set()
        for t in twins:
            
            peers_box = peers_box.union(peers[t])
            print("p",peers_box)
        box = twins[0]
        twindigits = values[box]
        print(twindigits)
        digit1 = twindigits[0]
        digit2 = twindigits[1]
        peers_box = peers[box] - set(twins)
        print(peers_box)
        for peer in peers_box:           
            values[peer] = values[peer].replace(digit1,'').replace(digit2,'') 
            print(values)
            print("nextline")
        return values
visited=[]
def naked_twins(values):
   
    twin_values = [box for box in values.keys() if len(values[box]) == 2 and box not in visited]
    if len(twin_values) < 2:
        return values
    for box in twin_values:
        digits = values[box]
        ti = [box]
        visited.append(box)
        tvalues = list(set(twin_values) - set(visited))
        candidates = [b for b in tvalues if b in peers[box] and values[b] == digits]
        for c in candidates:
            ti.append(c)
            visited.append(c)
        #if(len(ti) > 1):
         #   twin_instances.append(ti)
        if len(ti)> 1:
            break
    if len(ti)> 1:        
        values = eliminateNakedTwin(values,ti)
    return(naked_twins(values))

"""def naked_twins(values):
    Eliminate values using the naked twins strategy.
    Args:
        values(dict): a dictionary of the form {'box_name': '123456789', ...}

    Returns:
        the values dictionary with the naked twins eliminated from peers.
   

    # Find all instances of naked twins.If 2 cells in a row, column or the 3*3 square
    # can contain the same 2 digits only then they are called naked twins.
    #print(values.keys())
    #print(values)
    twin_instances=[]
    visited = []
    twin_values = [box for box in values.keys() if len(values[box]) == 2]
    print(twin_values)
    for box in twin_values:
        #find all the peers of box
        digits = values[box]
        ti = [box]
        visited.append(box)
        tvalues = list(set(twin_values) - set(visited))
        candidates = [b for b in tvalues if b in peers[box] and values[b] == digits]
        for c in candidates:
            ti.append(c)
        if(len(ti) > 1):
            twin_instances.append(ti)
        if len(twin_instances) == 1:
            
        digits = values[box]
        #print(box,digits)
        candidates = [b for b in twin_values if b in peers[box] and values[b] == digits and b not in visited]
        if(len(candidates) == 1):# and (candidates[0], not in visited:
            #found a new twin
            twin_instances.append((box,candidates[0]))
            visited.append(box)
            #print(box,candidates[0])   
    #print(len(twin_instances))      
    #print((twin_instances))
    
    # Eliminate the naked twins as possibilities for their peers
    #for box,box1 in twin_instances:
     for instance in twin_instances:
        box = instance[0]
        twindigits = values[box]
        print(twindigits)
        digit1 = twindigits[0]
        digit2 = twindigits[1]
        peers_box = peers[box] - set(instance)
        print(peers_box)
        for peer in peers_box:           
            values[peer] = values[peer].replace(digit1,'').replace(digit2,'') 
            
            #print("nextline")
    return values
"""

def eliminate(values):    
    solved_values = [box for box in values.keys() if len(values[box]) == 1]
    for box in solved_values:
        digit = values[box]
        for peer in peers[box]:
            values[peer] = values[peer].replace(digit,'')
    return values

def only_choice(values):
    for unit in unitlist:
        for digit in '123456789':
            dplaces = [box for box in unit if digit in values[box]]
            if len(dplaces) == 1:
                values[dplaces[0]] = digit
    return values

def reduce_puzzle(values):
    stalled = False
    while not stalled:
        # Check how many boxes have a determined value
        solved_values_before = len([box for box in values.keys() if len(values[box]) == 1])
        # Use the Eliminate Strategy
        values = eliminate(values)
        #use the naked twins eliminate strategy
        values = naked_twins(values)
        # Use the Only Choice Strategy
        values = only_choice(values)
        # Check how many boxes have a determined value, to compare
        solved_values_after = len([box for box in values.keys() if len(values[box]) == 1])
        # If no new values were added, stop the loop.
        stalled = solved_values_before == solved_values_after
        # Sanity check, return False if there is a box with zero available values:
        if len([box for box in values.keys() if len(values[box]) == 0]):
            return False
    return values

def search(values):
    values = reduce_puzzle(values)
    if(values is False):
        return False
    if all(len(values[s]) == 1 for s in boxes): 
        return values ## Solved!
    # Choose one of the unfilled squares with the fewest possibilities
    n,s = min((len(values[s]), s) for s in boxes if len(values[s]) > 1)
    #data = [(len(values[s]), s) for s in boxes if len(values[s]) > 1]
    #n,s = min(data, key = lambda t:t[0])

    # Now use recurrence to solve each one of the resulting sudokus, and 
    for value in values[s]:
        new_sudoku = values.copy()
        new_sudoku[s] = value
        attempt = search(new_sudoku)
        if attempt:
            return attempt
        
        
```

```python
from sklearn.datasets import load_iris
iris = load_iris()


average = np.mean(iris.data, axis=0) 
std     = np.std(iris.data, axis=0) 
range_  = range(np.shape(iris.data)[1])
```

```python
plt.subplot(1,2,1) # defines 1 row, 2 columns panel, activates figure 1
plt.title('Horizontal bars')
plt.barh(range_,average, color="r", xerr=std, alpha=0.4, align="center")
plt.yticks(range_, iris.feature_names)
plt.subplot(1,2,2) # defines 1 row 2 column panel, activates figure 2
plt.title('Vertical bars')
plt.bar(range_,average, color="b", yerr=std, alpha=0.4, align="center")
plt.xticks(range_, range_)
plt.show()
```
